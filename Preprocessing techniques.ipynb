# Data loading and Preprocessing 

```python
with open('assignment_data.txt', encoding='latin-1') as f:
    data = f.read()
```

```python
# removing the non alpha numerica characters
data_processed= re.sub('[^A-Za-z.]', ' ', data_processed)    

    # Remove all the special characters
data_processed = data_processed.replace(r'\W', ' ')
    # remove all single characters
data_processed = data_processed.replace(r'\s+[a-zA-Z]\s+', ' ')
    # Remove single characters from the start
data_processed = data_processed.replace(r'\^[a-zA-Z]\s+', ' ')
    # Substituting multiple spaces with single space
data_processed = data_processed.replace(r'\s+', ' ')
    # Removing prefixed 'b'
data_processed = data_processed.replace(r'^b\s+', '')
    # Remove leading, trailing spaces
data_processed = data_processed.strip()
```
